```python
def makeSearchWords(*pWord) :
    resultSearchWords = ''
    
    for word in pWord : 
        if resultSearchWords == '':
            resultSearchWords = word
        else :
            resultSearchWords += '+' + word
    
    return resultSearchWords
```

```python
# 검색어
searchWord = makeSearchWords('제네시스', 'GV80')
searchWord
```

```python
def makeExcludedWords(*pWord) :
    resultExcludeWords = ''
    
    for word in pWord : 
        resultExcludeWords += '+-' + word
    
    return resultExcludeWords
```

```python
# 다음 페이지 이동 버튼 존재 여부 확인 및 다음 페이지 이동
def getSearchResult(pCssSelector) :
    
    delay = 10 # 로딩 완료 대기 시간(초 단위)
    pageNum = 0 # debug용

    while True:    
        pageNum+=1
        print('page : ', pageNum)
            
        try :              
            getElement()
            
            # element 대입 시, 다음 페이지 이동 버튼(pCssSelector)이 없을 경우 찾는데 시간이 많이 소요되서 TimeoutException 발생
            element = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.CSS_SELECTOR, pCssSelector)))            
            
            # 다음 페이지 이동 버튼 있을 경우 Click Event 
            # Click Event 발생시, 마지막 페이지 이동되기 전에 오류 발생하는 경우 있어서 key event로 변경
            element.send_keys(Keys.ENTER)   
        
        except TimeoutException as toe:
            
            btnNext = driver.find_elements_by_css_selector(pCssSelector)
            
            # 해당 css selector가 없어서 못 읽어온 건지, 있는데 읽는데 오래 걸려서인지 확인
            print('TimeoutException : ', toe)  if existenceBtnNextPage(btnNext) else print('NoSuchElementException')            
            return   

        except Exception as e :
            print('예외 발생 :', e)
            return 
```

## 기사 본문 크롤링

```python
# 기사 본문 크롤링
def getConetent() :
    for idx in range(len(df)) :
        print('idx : ', idx)
        driver.get(df.loc[idx, 'link'])

        # 본문 크롤링
        content = driver.find_elements_by_css_selector('#articleBodyContents')
        
        # 위의 css selector가 없을 경우 pass
        if len(content) == 0:
            exclusionList.append(idx)
            pass
        else :        
            article = content[0].text        
            contentList.append(splitContent(article))   

            # 날짜 크롤링
            dateList.append(driver.find_element_by_css_selector('#main_content > div.article_header > div.article_info > div > span.t11').text[:10])        
```

```python
# 기사 문단별로 나누고, 필요없는 문단 제거
def splitContent(pArticle) : 
    
    subContentList = []
    
    # '\n\n' 기준으로 문단이 나뉘는 것으로 봄
    articleSp = pArticle.split('\n\n')
    
    for text in articleSp : 
        
        # 기사 건수가 많아서 본문 읽는데 시간 오래걸려 수집 본문 조건을 많이 검(전처리 최소화)
        if (text.find('▶') == -1) and (text.find('@') == -1) and (text.find('저작권자') == -1) and (text.find('무단전재') == -1) and (text.find('무단 전재') == -1) and (text.find('제공') == -1) and (text.find('사진') == -1) and (text.find('기자]') == -1) and (text.find('기자)') == -1) and (text.find('가기]') == -1) and (text.find('보기]') == -1) and (text.find('자료사진]') == -1) and (text.find('ⓒ') == -1) and (text.find('사진') == -1) and (text.find('서울=') == -1) and (text.find('고양=') == -1) and (text != '') and (text != ' '):
            text = text.replace('\n', ' ')
            
            # 비정형 데이터 분석에 불필요한 특수문제 제거
            text = re.sub('[▷△▽◇◆▲■●…;]', '', text)
            
            # 왼쪽 공백제거
            text = text.lstrip()
            
            subContentList.append(text)
            
    return contentList
```
