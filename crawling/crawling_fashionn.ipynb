## 패션엔 홈페이지에서 특정 브랜드 검색 후, 기사 제목 & 상세 페이지 주소 & 기사 작성일 크롤링
- https://www.fashionn.com/

```python
import pandas as pd
import sys
import datetime
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By   # 특정 요소 존재 여부 확인
from selenium.webdriver.support.ui import WebDriverWait # 웹 페이지 로딩 완료될 때까지 대기
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys # key event 발생 시킴

# Exception
from selenium.common.exceptions import NoSuchElementException
```

```python
path = '../driver/chromedriver.exe'
wUrl = 'https://www.fashionn.com/board/search_new.php?sel=all&search='
```

```python
def makeUrl(pUrl, pSearchWord) :
    return str(pUrl + pSearchWord)
```

```python
searhWord = {'셀린' : 'Ce',
             '디올'   : 'D',
             '샤넬'   : 'Ch',
             '로랑' : 'YSL'} # 입생로랑
```

```python
print(searhWord.keys())
print(searhWord.values())
```

### 브랜드 변경을 위한  index  변경

```python
searchWordIndex = 0
```

```python
brand = list(searhWord.keys())[searchWordIndex]
brand
```

```python
engBrand = list(searhWord.values())[searchWordIndex]
engBrand
```

```python
Url = wUrl
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
driver.get(makeUrl(Url, list(searhWord.keys())[searchWordIndex]))
```

### 검색한 페이지

```python
elements = driver.find_elements_by_css_selector('tbody > tr')
print('category :', elements[1].find_element_by_css_selector('td.author').text)
print('title :', elements[1].find_element_by_css_selector('td.tit > a').text)
print('date :', elements[1].find_element_by_css_selector('td.date').text)
print('link :', elements[1].find_element_by_css_selector('td.tit > a').get_attribute('href'))
```

```python
# 원하는 요소 정보 추출
def getElement() :
    elements = driver.find_elements_by_css_selector('tbody > tr')
    
    for idx in range(len(elements)) :
        if idx > 0 :
            categoryList.append(elements[idx].find_element_by_css_selector('td.author').text)
            titleList.append(elements[idx].find_element_by_css_selector('td.tit > a').text)
            dateList.append(elements[idx].find_element_by_css_selector('td.date').text)
            linkList.append(elements[idx].find_element_by_css_selector('td.tit > a').get_attribute('href'))
```

```python
# 크롤링 결과 반환
def getSearchResult(pCssSelector) :    

    for pageNum in range(1, sys.maxsize) :
        currentUrl = 'https://www.fashionn.com/board/search_new.php?page='+ str(pageNum) +'&sel=all&search='+ list(searhWord.keys())[searchWordIndex]
        driver.get(currentUrl)
        
        try : 
            
            # 다음 버튼 없을 경우 : NoSuchElementException
            driver.find_element_by_css_selector(pCssSelector)
            
            getElement()

            
        except NoSuchElementException as nse :
            print(nse)
            return
        
        except Exception as e :
            print('예외 발생 :', e)
            return                 
```
