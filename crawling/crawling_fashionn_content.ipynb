## 크롤링한 데이터 저장한 csv 파일을 읽고, 상세 페이지로 이동해서 기사 내용 크롤링
- https://www.fashionn.com/

```python
import pandas as pd
from selenium import webdriver
import datetime
```

### 크롤링할 데이터의 기본 정보

```python
searhWord = {'셀린' : 'Ce',
             '디올'   : 'D',
             '샤넬'   : 'Ch',
             '로랑' : 'YSL'} # 입생로랑
```

```python
searchWordIndex = 0
```

```python
brand = list(searhWord.keys())[searchWordIndex]
brand
```

```python
engBrand = list(searhWord.values())[searchWordIndex]
engBrand
```

```python
magazine = 'fashionn'
```

### csv 로드

```python
df = pd.read_csv('./data/'+ magazine +'_' + engBrand + '.csv', sep = '|')
```

```python
df.head()
```

### 웹 드라이버 실행

```python
path = '../driver/chromedriver.exe'
```

```python
url =  df.loc[0, 'link']
url
```

```python
driver = webdriver.Chrome(path)     # 크롬 브라우저 제어할 수 있는 객체, 웹브라우저 실행
driver
driver.get(url)
```

### p태그 읽어와서 list에 담음

```python
elements = driver.find_elements_by_css_selector('#content > div.text > div.board_view_wrap_new > div.view_body > p')
len(elements)
```

```python
contentList = []
```

```python
for element in elements :
    contentList.append(element.text)
```

```python
print(contentList)
```

### link 읽어서 contentList 만듬

```python
def getContentList(pDf, pStartIdx = 0) :
    
    for idx in range(pStartIdx, len(pDf)) :
        
        print('idx :', idx)

        # 해당 링크로 이동
        url = pDf.loc[idx, 'link']
        driver.get(url)
        
        # p태그 추출
        elements = driver.find_elements_by_css_selector('#content > div.text > div.board_view_wrap_new > div.view_body > p')

        # 초기화
        contentList = []

        for element in elements :
            contentList.append(element.text)
            
        colContentList.append(contentList)    
```

```python
colContentList = []
getContentList(df)
```

```python
# df의 길이와 colContenList 길이 비교하여 모든 페이지 크롤링 되었는지 확인
if len(df) == len(colContentList) :
    print(True)
else :
    print(False)
```

```python
df['content'] = colContentList
```
